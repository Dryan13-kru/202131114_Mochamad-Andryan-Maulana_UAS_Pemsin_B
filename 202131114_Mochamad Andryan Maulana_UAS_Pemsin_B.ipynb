```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming x_array is your data (NumPy array)
# Convert non-numeric values to NaN
def convert_to_numeric(column):
    return pd.to_numeric(column, errors='coerce')

x_array_numeric = np.apply_along_axis(convert_to_numeric, axis=0, arr=x_array)

# Replace NaN values with 0 using np.nan_to_num
x_array_numeric = np.nan_to_num(x_array_numeric, nan=0)

# Use the StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array_numeric)

# Scatter plot the scaled data
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], s=10)
plt.show()
```

```python
#CEK SUM OF SQUARE ERROR DARI TIAP PEMBAGIAN JUMLAH CLUSTER
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
    kmeans = KMeans(n_clusters = i, random_state = 0, n_init='auto')
    kmeans.fit(x_scaled)
    sse_ = kmeans.inertia_
    sse.append(sse_)
    print(i,sse_)
```

```python
plt.plot(index, sse)
plt.show()
```

```python
#membuat model K-Means
kmeans = KMeans (n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(x_scaled)
```

```python
#menemukan cluster pusatnya
kmeans.cluster_centers_
```

```python
#memvisualisasikan penyebaran datanya
output = plt.scatter(x_scaled[:,0],x_scaled[:,1], s=10, c = kmeans.labels_) #DATANYA

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="red", s=10) #centroid
plt.title("KMEANS Clustering Credit Card")
plt.xlabel("W1")
plt.ylabel("W2")
plt.colorbar(output)
plt.show()
```

```python
#evaluasi model
from sklearn.metrics import davies_bouldin_score
labels = kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(x_scaled, kmeans.labels_)
print(f"Silhouette Score: {silhouette_avg}")
```
