```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming x_array is your data (NumPy array)
# Convert non-numeric values to NaN
def convert_to_numeric(column):
    return pd.to_numeric(column, errors='coerce')

x_array_numeric = np.apply_along_axis(convert_to_numeric, axis=0, arr=x_array)

# Replace NaN values with 0 using np.nan_to_num
x_array_numeric = np.nan_to_num(x_array_numeric, nan=0)

# Use the StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array_numeric)

# Scatter plot the scaled data
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], s=10)
plt.show()
```

```python
#CEK SUM OF SQUARE ERROR DARI TIAP PEMBAGIAN JUMLAH CLUSTER
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
    kmeans = KMeans(n_clusters = i, random_state = 0, n_init='auto')
    kmeans.fit(x_scaled)
    sse_ = kmeans.inertia_
    sse.append(sse_)
    print(i,sse_)
```

```python
plt.plot(index, sse)
plt.show()
```

```python
#membuat model K-Means
kmeans = KMeans (n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(x_scaled)
```

```python
#menemukan cluster pusatnya
kmeans.cluster_centers_
```

```python
#memvisualisasikan penyebaran datanya
output = plt.scatter(x_scaled[:,0],x_scaled[:,1], s=10, c = kmeans.labels_) #DATANYA

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="red", s=10) #centroid
plt.title("KMEANS Clustering OLX Car Sell")
plt.xlabel("A")
plt.ylabel("B")
plt.colorbar(output)
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[]

# asumsi jumlah K berada di antara 1 sampai 10
for i in range(1,10):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss, color = 'blue')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# menentukan jumlah cluster
kmeans = KMeans(n_clusters=2)

# fitting input data
kmeans = kmeans.fit(X)

# mendapatkan cluster labels
labels = kmeans.predict(X)

# mendapatkan nilai centroid
centroids = kmeans.cluster_centers_
```

```python
# Contoh, jika labels diperoleh dari model K-Means
# Pastikan bahwa labels memiliki panjang yang sesuai dengan limited_dataset
labels = kmeans.labels_

# Tambahkan kolom 'Cluster' ke limited_dataset
limited_dataset['Cluster'] = labels

# Tampilkan limited_dataset dengan kolom Cluster
print(limited_dataset)
```

```python
kmeansmodel = KMeans(n_clusters= 2, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
y_kmeans # hasil clustering
```

```python
pred = y_kmeans
limited_dataset['Cluster'] = pred
limited_dataset['Cluster'].value_counts()
```

```python
# Visualizing all the clusters 
fig, ax = plt.subplots(figsize=(15, 7))
plt.scatter(X[pred == 0]['Year'], X[pred == 0]["KM's driven"], s=20, c='red', label='Cluster 0')
plt.scatter(X[pred == 1]['Year'], X[pred == 1]["KM's driven"], s=20, c='purple', label='Cluster 1')
plt.scatter(X[pred == 2]['Year'], X[pred == 2]["KM's driven"], s=20, c='green', label='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', label='Cluster Centers', marker='*')
plt.title('Clusters of OLX Car Sell')
plt.xlabel('Year')
plt.ylabel("KM's driven")
plt.legend(loc='best')
plt.show()
```

Berikut penjelasan singkat untuk elemen-elemen plot:

Scatter Plot (Klaster):

Titik-titik merah (Cluster 0): Mewakili data yang terkait dengan klaster pertama.  
Titik-titik ungu (Cluster 1): Mewakili data yang terkait dengan klaster kedua.            
Titik-titik hijau (Cluster 2): Mewakili data yang terkait dengan klaster ketiga.

Cluster Centers (Bintang Hitam):

Menunjukkan pusat dari masing-masing klaster yang diidentifikasi oleh algoritma K-Means.
Label dan Axis:

Judul plot ("Clusters of OLX Car Sell") memberikan gambaran umum tentang konten plot.
Label sumbu x ("Year") dan label sumbu y ("KM's driven") memberikan informasi tentang variabel yang direpresentasikan oleh sumbu-sumbu tersebut.
